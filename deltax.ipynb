# Variazione di $\Delta{x}$

## Importazione delle librerie necessarie

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Scelta del dominio

```python
x=np.linspace(-2,2,201)
#print(x)
```

## Definizione della funzione da analizzare

```python
def f(x):
    return(x**5)
```

## Definizione della funzione pendenza

```python
def m(x,dx):
    return((f(x+dx)-f(x))/dx)
```

## Analisi

```python
plt.figure(figsize=(9, 6))
plt.plot(x, f(x))
for dx in [1, 0.5, 0.1, 1e-9]:
    plt.plot(x, m(x,dx))   
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('f(x) e pendenze per diversi dx')
plt.grid(True)
plt.ylim(-1, 2)
plt.show()
```

```python
dx=np.linspace(2,1e-9,201)

plt.figure(figsize=(9, 6))
plt.plot(dx,m(0,dx))
plt.xlabel('dx')
plt.ylabel('m(0,dx)')
plt.title('Pendenza in funzione di dx')
plt.grid(True)
plt.ylim(-1, 2)
plt.show()
```
