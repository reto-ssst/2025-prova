# Integrazione numerica in Python

```python
import numpy as np
from numpy import pi
import matplotlib.pyplot as plt
```

```python
def f(x):
    return np.sin(x)
```

```python
x=np.linspace(0,pi/2,5)
y=f(x)
plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('f(x) = sin(x)')
plt.title('Plot of f(x) = sin(x)')
plt.grid()
plt.show()
```

```python
def int_num(f,xi,xf,n):
    dx=(xf-xi)/n
    A=0
    x=xi
    for i in range(n):
        A=A + f(x)*dx  
        print(i,x,A)
        x = x + dx
    return A
```

```python
Area = int_num(f,0,pi/2,4)
```

```python
def trapezi(f,xi,xf,n):
    dx=(xf-xi)/n
    yt=f(xi)/2
    #print(0,xi,yt)
    x=xi+dx
    for i in range(1,n):
        yt=yt + f(x)
        #print(i,x,yt)
        x = x + dx
    yt=yt + f(xf)/2
    A=yt*dx
   # print(n,xf,yt)
    return A    
```

```python
Area_trap = trapezi(f,0,pi/2,2)
print("Area con il metodo dei trapezi:", Area_trap)
```

```python
n=2
for i in range(4):
    Area_trap = trapezi(f,0,pi/2,n)
    print("n =", n, " A =", Area_trap)
    n=n*2
```
