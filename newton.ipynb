# Determinazione degli zeri di una funzione con il metodo di Newton

## Importazione delle librerie necessarie

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Definizione della funzione

```python
def f(x):
    return(x**3+x**2-5*x+1)
```

## Pendenza

```python
def m(x):
    dx=1e-6
    return((f(x+dx)-f(x))/dx)
```

## Grafico

```python
x=np.linspace(-4,3,100)
```

```python
plt.figure(figsize=(9, 6))
plt.plot(x, f(x))
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Grafico di f(x)')
plt.grid(True)
plt.ylim(-10, 10)
plt.show()
```

## Determinazione degli zeri

Scelgo il valore iniziale $x_0$

```python
x0=-3
```

Calcolo $f(x_0)$ e $m(x_0)$, la pendenzza in $x_0$ 

```python
print('f(x0)=',f(x0))
print('m(x0)=',m(x0))
```

Cerco lo zero

```python
xn=x0
while abs(f(xn))>1e-6:
    xn=xn-f(xn)/m(xn)
    print(xn,f(xn))

print('\n Radice trovata in x=',xn)
```
